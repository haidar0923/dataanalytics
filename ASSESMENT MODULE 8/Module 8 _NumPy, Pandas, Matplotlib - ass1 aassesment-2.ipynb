```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "historical_automobile_sales.csv"  # Ensure the correct file path
df = pd.read_csv(file_path)
```

```python
# Set Seaborn theme
sns.set_theme(style="darkgrid")
```

```python
### Q1: Line chart showing automobile sales fluctuations over the years
plt.figure(figsize=(12, 6))
sales_per_year = df.groupby("Year")["Automobile_Sales"].sum()  # Aggregate sales by year
sales_per_year.plot(marker="o", linestyle="-", color="b")
plt.title("Automobile Sales Fluctuations Over the Years", fontsize=14)
plt.xlabel("Year")
plt.ylabel("total Automobile Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
### Q2: Multiple Line Chart for Different Vehicle Types
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Year", y="Automobile_Sales", hue="Vehicle_Type", marker="o")

plt.title("Automobile Sales Trends by Vehicle Type", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Total Automobile Sales")
plt.legend(title="Vehicle Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="Year", y="Automobile_Sales", hue="Recession")

plt.title("Sales Trend: Recession vs. Non-Recession", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Automobile Sales")
plt.xticks(rotation=45)
plt.legend(title="Recession (0 = No, 1 = Yes)")
plt.grid(True)

plt.show()
```
